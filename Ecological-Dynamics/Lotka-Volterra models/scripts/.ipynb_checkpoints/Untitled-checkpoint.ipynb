# Modelling mutualistic interactions in chemically-mediated models

Modelling mutualistic interactions is challenging because modelled growth is often unbounded, so biologically infeasible.
Constructing chemically-mediated models is also challenging because there are many potential mechanisms that could drive chemically-mediated species interactions. 

In this document, I will discuss the strengths and weakness of candidate models of chemically-mediated mutualisms, specifically cross-feeding. All models will have 2 species that can cross-feed each other through leached metabolites, or compete for a single external resource.

```python
import numpy as np
import seaborn as sns
import matplotlib as mpl
from matplotlib import pyplot as plt
from scipy.integrate import solve_ivp
```

## 1. The simplest model

This model is a simple consumer-resource model with a fixed growth rate, resource consumption rate and production rate.
There is a single leach metabolite pool that species can contribute to and feed off from.

$$ \frac{\text{d}N_i}{\text{d}t} = \underbrace{N_i\sum_{c=1}^C g_{i,c}\beta_{i,c}R_c}_{\text{growth}} - \underbrace{\alpha_iN_i}_{\text{death}}, \ 
\frac{\text{d}R_\text{ext}}{\text{d}t} = \text{influx} - \underbrace{R_{ext}\sum_{i=1}^2\beta_{i,ext}N_i}_{\text{consumption}}, \
\frac{\text{d}R_\text{leach}}{\text{d}t} =  \underbrace{\sum_{i=1}^2p_iN_i}_{\text{production}} - \underbrace{R_{leach}\sum_{i=1}^2\beta_{i,leach}N_i}_{\text{consumption}}$$

$N_i $ is the biomass of species $i$. $R_{ext}$ is the concentration of the external nutrient. $R_{leach}$ is the concentration of the leached metabolite.

$\beta_{i,c}$ is the consumption of resource $c$ by species $i$. $p_{i}$ is the production of the leached metabolite by species $i$. $g_{i,c}$ is the growth of species $i$ on resource $c$. $\alpha_i$ is the death rate of species $i$.

```python
def dSRdt(t,var,
          growth,death,
          influx,consumption,production):
    
    
    S = var[:2]
    R_e = var[2]
    R_l = var[3]
    
    dS = np.sum(growth * consumption * np.array([R_e,R_l]), axis=1)*S - death*S
    
    dR_e = influx - np.sum(consumption[:,0] * R_e * S)
    dR_l = np.sum(production * S) - np.sum(consumption[:,1] * R_l * S)
    
    dSR = np.concatenate([dS,np.array([dR_e,dR_l])])
    
    return dSR
```

### Issues with the model

- Leached metabolite production does not depend on resource availability or consumption.
- A species can feed off its own waste products/leached metabolites, so it can grow unbounded in monoculture or co-culture.

#### Monocultural growth

```python
growth = np.array([[0.001,0.1],
                   [0,0]])
death = np.array([0.01,0])

influx = 0
production = np.array([1,0])
consumption = np.array([[0.1,1],
                        [0,0]])

t_end = 100
initial_abundance = np.array([0.1,0,5,0])

simulation = solve_ivp(dSRdt,[0,t_end],initial_abundance,
                 args=(growth,death,influx,consumption,production),
                 method='RK45')
```

```python
plt.plot(simulation.t,simulation.y[0,:].T)
plt.xlabel('time',fontsize=18)
plt.ylabel('species abundance',fontsize=18)
```

## 2. Adding multiple leached metabolites

This model expands on the simplest model by assuming each species produces a different secondary metabolite. Species cannot grow off their own waste products. In addition, production of the leached metabolite is dependent on consumption of other resources.

$$ \frac{\text{d}N_i}{\text{d}t} = \underbrace{N_i\sum_{c=1}^C g_{i,c}\beta_{i,c}R_c}_{\text{growth}} - \underbrace{\alpha_iN_i}_{\text{death}}, \ 
\frac{\text{d}R_\text{external}}{\text{d}t} = \text{influx} - \underbrace{R_{external}\sum_{i=1}^2\beta_{i,c}N_i}_{\text{consumption}}, \
\frac{\text{d}R_\text{l}}{\text{d}t} = \underbrace{\sum_{i=1}^2p_iN_i\sum_{m=1(m \neq l)}^M\beta_{i,m}R_m}_{\text{production}} - \underbrace{R_{l}\sum_{i=1}^2\beta_{i,l}N_i}_{\text{consumption}}$$

$N_i $ is the biomass of species $i$. $R_{ext}$ is the concentration of the external nutrient. $R_{l}$ is the concentration of  leached metabolite $l$ - there are 2 in total.

```python
def dSRdt(t,var,
          growth,death,
          influx,consumption,production):
    
    S = var[:2]
    R = var[2:]
    
    dS = np.sum(growth * consumption * R, axis=1)*S - death*S
    
    def production_rate(i):
        
        R_remove_i = np.delete(R,i)
        consump_remove_i = np.delete(consumption,i,axis=1)
         
        R_i_produced = production[:,i] * np.sum(consump_remove_i * R_remove_i,axis=1)
         
        return R_i_produced
    
    R_produced = np.array([production_rate(i) for i in range(len(R))])
    R_consumed = consumption * R
    
    production_consumption = \
        (np.matmul(R_produced, S.reshape(len(S),1)) - \
         np.matmul(R_consumed.T, S.reshape(len(S),1))).reshape(R.shape)
     
    dR = influx + production_consumption
    
    dSR = np.concatenate([dS,dR])
    
    return dSR
```

### Issues with the model

- This model will still have unbounded growth in co-culture.

#### Coculture growth

```python
growth = np.array([[0.001,0,0.1],
                   [0.002,0.05,0]])
death = np.array([0.01,0.01])

influx = np.zeros(3)
production = np.array([[0,1,0],
                        [0,0,1]])
consumption = np.array([[0.1,0,1],
                        [0.1,1,0]])

t_end = 70
initial_abundance = np.array([0.1,0.1,5,0,0])

simulation = solve_ivp(dSRdt,[0,t_end],initial_abundance,
                 args=(growth,death,influx,consumption,production),
                 method='RK45')
```

```python
plt.plot(simulation.t,simulation.y[:2,:].T)
plt.xlabel('time',fontsize=18)
plt.ylabel('species abundance',fontsize=18)
plt.legend(['Species 1', 'Species 2'],fontsize=12)
```

## 3. Adding metabolic constraints to the model

In order to prevent unbounded growth under no influx of external resources, I propose that a species cannot leach metabolites without the external nutrient. In this case, the external nutrient is some media that lacks one key nutrient for each species. Therefore, species need to cross-feed to grow, but cannot grow in the absence of external nutrient.

The model is the same as (2), but I have expanded the $R_l$ production term to make it dependent on the consumption of $R_external$.

$$ \frac{\text{d}R_\text{l}}{\text{d}t} = \underbrace{\sum_{i=1}^2p_iN_i\sum_{m=1(\neq l)}^M(\beta_{i,m}R_m)(\beta_{i,\text{ext}}R_{\text{ext}})}_{\text{production}} - \underbrace{R_{l}\sum_{i=1}^2\beta_{i,l}N_i}_{\text{consumption}}$$

```python
def dSRdt(t,var,
          growth,death,
          influx,consumption,production):
    
    S = var[:2]
    R = var[2:]
    
    dS = np.sum(growth * consumption * R, axis=1)*S - death*S
    
    def production_rate(i):
        
        R_remove_i = np.delete(R,i)
        consump_remove_i = np.delete(consumption,i,axis=1)
         
        R_i_produced = production[:,i] * \
            (R[0] * consumption[:,0][:, np.newaxis])[:,0] * \
                np.sum(consump_remove_i * R_remove_i,axis=1)
         
        return R_i_produced
    
    R_produced = np.array([production_rate(i) for i in range(len(R))])
    R_consumed = consumption * R
    
    production_consumption = \
        (np.matmul(R_produced, S.reshape(len(S),1)) - \
         np.matmul(R_consumed.T, S.reshape(len(S),1))).reshape(R.shape)
     
    dR = influx + production_consumption
    
    dSR = np.concatenate([dS,dR])
    
    return dSR
```

When the death rate is 0, the model achieves steady state when there is no influx of nutrients. This because species' relative growth rate declines to 0 once the $R_{\text{ext}}$ is used up.

```python
growth = np.array([[0.001,0,0.1],
                   [0.002,0.05,0]])
death = np.array([0.0,0.0])

influx = np.zeros(3)
production = np.array([[0,1,0],
                        [0,0,1]])
consumption = np.array([[0.1,0,1],
                        [0.1,1,0]])

t_end = 150
initial_abundance = np.array([0.1,0.1,5,0,0])

simulation = solve_ivp(dSRdt,[0,t_end],initial_abundance,
                 args=(growth,death,influx,consumption,production),
                 method='RK45')
```

## 4. Adding logistic growth

To bound exponential growth, one can add a carrying capacity or self-inhibition. Although this is commonly done in abstract model, it is not clear what mechanism would be driving self-inhibition in our chemically-mediated model.

The model is the same as (3) with a modification to species growth rate:

$$ \frac{\text{d}N_i}{\text{d}t} =  N_i(1-N_i)\underbrace{\sum_{c=1}^C g_{i,c}\beta_{i,c}R_c}_{\text{effective growth rate}} $$

```python
def dSRdt(t,var,
          growth,
          influx,consumption,production):
    
    S = var[:2]
    R = var[2:]
    
    dS = np.sum(growth * consumption * R, axis=1) * S * (1 - S)
    
    def production_rate(i):
        
        R_remove_i = np.delete(R,i)
        consump_remove_i = np.delete(consumption,i,axis=1)
         
        R_i_produced = production[:,i] * \
            (R[0] * consumption[:,0][:, np.newaxis])[:,0] * \
                np.sum(consump_remove_i * R_remove_i,axis=1)
         
        return R_i_produced
    
    R_produced = np.array([production_rate(i) for i in range(len(R))])
    R_consumed = consumption * R
    
    production_consumption = \
        (np.matmul(R_produced, S.reshape(len(S),1)) - \
         np.matmul(R_consumed.T, S.reshape(len(S),1))).reshape(R.shape)
     
    dR = influx + production_consumption
    
    dSR = np.concatenate([dS,dR])
    
    return dSR
```

Similarly to (3), microbial growth curves initially have exponential growth, then transition to a linear growth phase.

```python
growth = np.array([[0.001,0,0.1],
                   [0.002,0.05,0]])

influx = np.array([0.1,0,0])
production = np.array([[0,1,0],
                        [0,0,1]])
consumption = np.array([[0.1,0,1],
                        [0.1,1,0]])

t_end = 150
initial_abundance = np.array([0.1,0.1,5,0,0])

simulation = solve_ivp(dSRdt,[0,t_end],initial_abundance,
                 args=(growth,influx,consumption,production),
                 method='RK45')
```

```python
plt.plot(simulation.t,simulation.y[:2,:].T)
plt.xlabel('time',fontsize=18)
plt.ylabel('species abundance',fontsize=18)
plt.legend(['Species 1', 'Species 2'],fontsize=12)
```

If the carrying capacity is independent of species growth rate, so does not depend on nutrient availability, the model reaches a steady state under an influx of nutrients. However, the growth curve is unusual compared to standard microbial growth curves - rather than exponential growth to a stationary point, the model oscillates until it achieves steady state.

```python
def dSRdt(t,var,
          growth,
          influx,consumption,production):
    
    S = var[:2]
    R = var[2:]
    
    dS = np.sum(growth * consumption * R, axis=1) * S  - S**2
    
    def production_rate(i):
        
        R_remove_i = np.delete(R,i)
        consump_remove_i = np.delete(consumption,i,axis=1)
         
        R_i_produced = production[:,i] * \
            (R[0] * consumption[:,0][:, np.newaxis])[:,0] * \
                np.sum(consump_remove_i * R_remove_i,axis=1)
         
        return R_i_produced
    
    R_produced = np.array([production_rate(i) for i in range(len(R))])
    R_consumed = consumption * R
    
    production_consumption = \
        (np.matmul(R_produced, S.reshape(len(S),1)) - \
         np.matmul(R_consumed.T, S.reshape(len(S),1))).reshape(R.shape)
     
    dR = influx + production_consumption
    
    dSR = np.concatenate([dS,dR])
    
    return dSR
```

## An alternative model with saturating growth rates

Many chemically-mediated models have consumption rates that saturate a high resource concentration. These models are more biologically realistic and prone to stability than models with non-saturating resource uptake.

In this model, we do not constrain metabolite leaching with the presence of an external nutrient source. However, we are able to achieve similar dynamics to models (3) and (4).

```python
def dCRdt(t,var,
          growth,max_consumption,half_saturation,production,
          influx):
     
    S = var[:2]
    R = var[2:]
    
    dS = np.sum(growth * ((max_consumption*R)/(R+half_saturation)),axis=1) * S * (1 - S)
    
    def production_rate(i):
        
        R_remove_i = np.delete(R,i)
        max_consump_remove_i = np.delete(max_consumption,i,axis=1)
        half_sat_remove_i = np.delete(half_saturation,i,axis=1)
        
        R_i_produced = production[:,i] * \
        np.sum(max_consump_remove_i * R_remove_i/(R_remove_i+half_sat_remove_i),axis=1)
         
        return R_i_produced
    
    R_produced = np.array([production_rate(i) for i in range(len(R))])
        
    R_consumed = max_consumption * (R/(R+half_saturation))
    
    production_consumption = \
        (np.matmul(R_produced, S.reshape(len(S),1)) - \
         np.matmul(R_consumed.T, S.reshape(len(S),1))).reshape(R.shape)
        
    dR = influx + production_consumption
    
    dSR = np.concatenate([dS, dR])
    
    return dSR

t_end = 70
initial_abundance = np.array([0.1,0.1,5,0,0])
growth = np.array([[0.001,0,1],[0.001,1,0]])
half_saturation = np.ones((2,3))
max_consumption = np.array([[0.1,0,1],[0.1,1,0]])
production = np.array([[0,1,0],[0,0,1]])
influx = np.array([0.5,0,0])

simulation = solve_ivp(dCRdt,[0,t_end],initial_abundance,
                 args=(growth,max_consumption,half_saturation,
                       production,influx),method='RK45')

plt.plot(simulation.t,simulation.y[:2,:].T)
plt.xlabel('time',fontsize=18)
plt.ylabel('species abundance',fontsize=18)
plt.legend(['Species 1', 'Species 2'],fontsize=12)
```

$$ \frac{\text{d}N_i}{\text{d}t} = \underbrace{N_i\sum_{c=1}^C g_{i,c}\beta_{i,c}R_c}_{\text{growth}} - \underbrace{\alpha_iN_i}_{\text{death}}, \ 
\frac{\text{d}R_\text{ext}}{\text{d}t} = \text{influx} - \underbrace{R_{ext}\sum_{i=1}^2\beta_{i,ext}N_i}_{\text{consumption}}, \
\frac{\text{d}R_\text{leach}}{\text{d}t} =  \underbrace{\sum_{i=1}^2p_iN_i \sum_{c=1(\neq l)}^M\beta_{i,c}R_c}_{\text{production}} - \underbrace{R_{leach}\sum_{i=1}^2\beta_{i,leach}N_i}_{\text{consumption}} $$

$$ \text{where } \beta_{i,c} = \frac{V_{i,c}}{k_{i,c} + R_c}. $$

$V_{i,c}$ is the maximum consumption of resource $c$ by species $i$. $k_{i,c}$ is the half-saturation rate.

```python
def dSRdt(t,var,
          growth,death,
          max_consumption,half_saturation,production,
          influx):
     
    S = var[0:2]
    R = var[2:]
    
    dS = np.sum(growth * ((max_consumption*R)/(R+half_saturation)),axis=1)*S - death*S
    
    def production_rate(i):
        
        R_remove_i = np.delete(R,i)
        max_consump_remove_i = np.delete(max_consumption,i,axis=1)
        half_sat_remove_i = np.delete(half_saturation,i,axis=1)
        
        R_i_produced = production[:,i] * \
        np.sum(max_consump_remove_i * R_remove_i/(R_remove_i+half_sat_remove_i),axis=1)
         
        return R_i_produced
    
    R_produced = np.array([production_rate(i) for i in range(len(R))])
        
    R_consumed = max_consumption * (R/(R+half_saturation))
    
    production_consumption = \
        (np.matmul(R_produced, S.reshape(len(S),1)) - \
         np.matmul(R_consumed.T, S.reshape(len(S),1))).reshape(R.shape)
        
    dR = influx + production_consumption
    
    dSR = np.concatenate([dS, dR])
    
    return dSR
```

If we expand this model so that metabolite leaching is dependent on external resources, we get very similar model behaviour.

```python
def dSRdt(t,var,
          growth,
          max_consumption,half_saturation,production,
          influx):
     
    S = var[0:2]
    R = var[2:]
    
    dS = np.sum(growth * ((max_consumption*R)/(R+half_saturation)),axis=1) * S * (1 - S)
    
    def production_rate(i):
        
        R_remove_i = np.delete(R,i)
        max_consump_remove_i = np.delete(max_consumption,i,axis=1)
        half_sat_remove_i = np.delete(half_saturation,i,axis=1)
        
        R_i_produced = production[:,i] * \
            np.sum(max_consump_remove_i * \
                   (R[0]/(R[0] + half_saturation[:,0])) * \
                       (R_remove_i/(R_remove_i+half_sat_remove_i)),axis=1)
         
        return R_i_produced
    
    R_produced = np.array([production_rate(i) for i in range(len(R))])
        
    R_consumed = max_consumption * (R/(R+half_saturation))
    
    production_consumption = \
        (np.matmul(R_produced, S.reshape(len(S),1)) - \
         np.matmul(R_consumed.T, S.reshape(len(S),1))).reshape(R.shape)
        
    dR = influx + production_consumption
    
    dSR = np.concatenate([dS, dR])
    
    return dSR
```

```python
growth = np.array([[0.001,0,0.1],
                   [0.002,0.05,0]])

influx = np.array([0.1,0,0])
production = np.array([[0,1,0],
                       [0,0,1]])
max_consumption = np.array([[0.1,0,1],
                            [0.1,1,0]])
half_saturation = np.ones((2,3))

t_end = 250
initial_abundance = np.array([0.1,0.1,5,0,0])

simulation = solve_ivp(dSRdt,[0,t_end],initial_abundance,
                 args=(growth,max_consumption,half_saturation,
                       production,influx),method='RK45')

plt.plot(simulation.t,simulation.y[:2].T)
```

## A model with logistic and non-saturating growth and no dependence on the external nutrients, can achieve steady state.

```python
def dSRdt(t,var,
          growth,
          consumption,production,influx):
     
    S = var[0:2]
    R = var[2:]
    
    dS = np.sum(growth * consumption * R,axis=1) * S * (1 - S)
    
    def production_rate(i):
        
        R_remove_i = np.delete(R,i)
        consump_remove_i = np.delete(consumption,i,axis=1)
        
        R_i_produced = production[:,i] * \
        np.sum(consump_remove_i * R_remove_i,axis=1)
         
        return R_i_produced
    
    R_produced = np.array([production_rate(i) for i in range(len(R))])
        
    R_consumed = consumption * R
    
    production_consumption = \
        (np.matmul(R_produced, S.reshape(len(S),1)) - \
         np.matmul(R_consumed.T, S.reshape(len(S),1))).reshape(R.shape)
        
    dR = influx + production_consumption
    
    dSR = np.concatenate([dS, dR])
    
    return dSR
```

```python
growth = np.array([[0.001,0,0.1],
                   [0.002,0.05,0]])

influx = np.array([0.1,0,0])
production = np.array([[0,1,0],
                       [0,0,1]])
consumption = np.array([[0.1,0,1],
                        [0.1,1,0]])

t_end = 100
initial_abundance = np.array([0.1,0.1,0,0,0])

simulation = solve_ivp(dSRdt,[0,t_end],initial_abundance,
                 args=(growth,consumption,production,influx),
                       method='RK45')

plt.plot(simulation.t,simulation.y[:2].T)
```

The problem is, of course, species can grow in the absence of any nutrients.

## Summary

We have 2 candidate models: (1) the non-saturating growth model where metabolite leaching depends on an external resource; (2) a saturating growth model where metabolite leaching does not depend on an external resource.

### Candidate model 1

$$ \frac{\text{d}N_i}{\text{d}t} = \underbrace{N_i\sum_{c=1}^C g_{i,c}\beta_{i,c}R_c}_{\text{growth}}, \ 
\frac{\text{d}R_\text{external}}{\text{d}t} = \text{influx} - \underbrace{R_{external}\sum_{i=1}^S\beta_{i,c}N_i}_{\text{consumption}}, \
\frac{\text{d}R_\text{l}}{\text{d}t} = \underbrace{\sum_{i=1}^Sp_iN_i(\beta_{i,\text{ext}}R_{\text{ext}})\sum_{m=1(m \neq l)}^M(\beta_{i,m}R_m)}_{\text{production}} - \underbrace{R_{l}\sum_{i=1}^S\beta_{i,l}N_i}_{\text{consumption}} $$

$N_i $ is the biomass of species $i$. $R_{ext}$ is the concentration of the external nutrient. $R_{l}$ is the concentration of  leached metabolite $l$.

$\beta_{i,c}$ is the consumption of resource $c$ by species $i$. $p_{i}$ is the production of the leached metabolite by species $i$. $g_{i,c}$ is the growth of species $i$ on resource $c$.

If we feel like it, we can add logistic growth, but this does not make a significant difference to mutualism dynamics. However, it will probably make competitive interactions more realistic.

$$ \frac{\text{d}N_i}{\text{d}t} =  G_iN_i\left(1-N_i\right), \text{ where } G_i = \sum_{c=1}^C g_{i,c}\beta_{i,c}R_c. $$

### Candidate model 2

$$ \frac{\text{d}N_i}{\text{d}t} = G_iN_i\left(1-N_i\right), \ 
\frac{\text{d}R_\text{ext}}{\text{d}t} = \text{influx} - \underbrace{R_{ext}\sum_{i=1}^S\beta_{i,ext}N_i}_{\text{consumption}}, \
\frac{\text{d}R_\text{leach}}{\text{d}t} =  \underbrace{\sum_{i=1}^Sp_iN_i \sum_{m=1(\neq l)}^M\beta_{i,m}R_m}_{\text{production}} - \underbrace{R_{leach}\sum_{i=1}^S\beta_{i,leach}N_i}_{\text{consumption}} $$

$$ \text{where } G_i = \sum_{c=1}^C g_{i,c}\beta_{i,c}R_c 
\text{ and }
\beta_{i,c} = \frac{V_{i,c}}{k_{i,c} + R_c}. $$

$N_i $ is the biomass of species $i$. $R_{ext}$ is the concentration of the external nutrient. $R_{l}$ is the concentration of  leached metabolite $l$.

$\beta_{i,c}$ is the saturating consumption of resource $c$ by species $i$. This term breaks down into $V_{i,c}$, the maximum consumption of resource $c$ by species $i$, and $k_{i,c}$ the half-saturation rate. $p_{i}$ is the production of the leached metabolite by species $i$. $g_{i,c}$ is the growth of species $i$ on resource $c$.

If we feel like it, we can add dependence on the external nutrient to metabolite leaching, but this does not make a significant difference to model outputs.

$$ \underbrace{\sum_{i=1}^S p_iN_i(\beta_{i,\text{ext}}R_{\text{ext}})\sum_{m=1(\neq l)}^M(\beta_{i,m}R_m)}_{\text{production}} $$
